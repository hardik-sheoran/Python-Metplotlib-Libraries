```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
plt.scatter(day,no)
plt.title("Scatter Plot")
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No.")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ['r','y','g','b','r','g','r']
size = [100,20,40,50,20,10,70]
plt.scatter(day,no,color = c,s = size)
plt.title("Scatter Plot")
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No.")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ['r','y','g','b','r','g','r']
size = [100,20,40,50,20,10,70]
al = [0.1,0.2,1,0.4,0.5,0.6,1]
plt.scatter(day,no,color = c,s = size,alpha=al)
plt.title("Scatter Plot")
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No.")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ['r','y','g','b','r','g','r']
size = [100,20,40,50,20,10,70]
al = [0.1,0.2,1,0.4,0.5,0.6,1]
plt.scatter(day,no,color = c,s = size,alpha=al,marker = "*")
plt.title("Scatter Plot")
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No.")
plt.show()
```

```python
day = [1,2,3,4,5,6,7]
no = [2,3,1,4,5,3,6]
c = ['r','y','g','b','r','g','r']
size = [100,20,40,50,20,10,70]
al = [0.1,0.2,1,0.4,0.5,0.6,1]
plt.scatter(day,no,color = c,s = size,alpha=al,marker = "*",edgecolors='m',linewidths=3)
plt.title("Scatter Plot")
plt.xlabel("Day", fontsize = 10)
plt.ylabel("No.")
plt.show()
```
