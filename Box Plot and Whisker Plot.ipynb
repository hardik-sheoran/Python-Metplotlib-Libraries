```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,notch = True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,vert=False)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,widths = 0.8)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,labels=['Python'])
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,patch_artist=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,showmeans=True)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120] # The dot in know as outler

plt.boxplot(x)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,whis = 3)
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,sym = "g+")
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,showmeans=True,sym = " ")
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,showmeans=True,sym = "g+",boxprops=(dict(color='r')),capprops=dict(color='m'))
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x,boxprops=(dict(color='r')),capprops=dict(color='m'),whiskerprops=dict(color='y'))
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]

plt.boxplot(x,flierprops=dict(markeredgecolor='y'))
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
x1 = [10,20,30,40,50,60,70]

plt.boxplot(x,flierprops=dict(markeredgecolor='y'))
plt.boxplot(x1)

plt.show()
```

```python
x = [10,20,30,40,50,60,70,]
x1 = [10,20,40,50,30,60,90]

y = [x,x1]

plt.boxplot(y,labels = ["Python","C++"])
plt.show()
```
