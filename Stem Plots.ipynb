```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=':')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=':',markerfmt='r^')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',bottom=1) # bottom = 9, make the graph upside down.
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g')
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g',label = "python")

plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g',use_line_collection=False)

plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,6,4,3]

plt.stem(x,y,linefmt=':',markerfmt='ro',basefmt='g',use_line_collection=False,orientation='horizontal')

plt.legend()
plt.show()
```
