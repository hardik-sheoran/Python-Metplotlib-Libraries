```python
import numpy as np

x = ['Python','C','c++','JAVA']
y = [85,70,60,82]
z = [20,30,40,50]

l = np.arange(len(x))
width = 0.2
l1 = [j+width for j in l]

plt.xlabel('language',fontsize = 15)
plt.ylabel('Popularity',fontsize = 15)
plt.title("Python",fontsize = 15)

plt.bar(l,y,width,color = 'r',label="Popularity")
plt.bar(l1,z,width,color = 'y',label="Popularity1")
plt.legend()

plt.show()
```
