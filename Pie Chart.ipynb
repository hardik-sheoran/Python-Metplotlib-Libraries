```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40]
y = ["C","C++","Java",'Python']

plt.pie(x)
plt.show()
```

## Donut Pie Chart

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["C","C++","Java",'Python']

ex = [0.0,0.0,0.1,0.0]
c = ['r','g','b','y']

plt.pie(x,labels=y,radius = 1.5)
plt.pie(x1,radius = 1,colors = c)

plt.show()
```

## Ring Chart

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["C","C++","Java",'Python']

ex = [0.0,0.0,0.1,0.0]
c = ['r','g','b','y']

plt.pie(x,labels=y,radius = 1.5)
plt.pie(x1,radius = 1,colors = "w")

plt.show()
```

```python
x = [10,20,30,40]
x1 = [40,30,20,10]
y = ["C","C++","Java",'Python']

ex = [0.0,0.0,0.1,0.0]
c = ['r','g','b','y']

plt.pie(x,labels=y,radius = 1.5)
cr = plt.Circle(xy=(0,0),radius = 1,facecolor = "w")
plt.gca().add_artist(cr)

plt.show()
```
